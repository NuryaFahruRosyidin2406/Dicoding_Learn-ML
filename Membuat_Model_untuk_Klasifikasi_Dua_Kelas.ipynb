```python
import pandas as pd

url = 'https://github.com/NuryaFahruRosyidin2406/Dicoding_Learn-ML/raw/main/Dataset/Membuat%20Model%20untuk%20Klasifikasi%20Dua%20Kelas/citrus.csv'
# df = pd.read_csv('citrus.csv')
df = pd.read_csv(url, encoding='utf-8')
```

```python
df.info()
```

```python
df.head(10000)
```

```python
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
```

```python
df.head(10000)
```

```python
dataset = df.values
dataset
```

```python
# pilih 5 kolom terakhir sebagai atribut
X = dataset[:,1:6]
X
# bilangan sebelum koma untuk memilih baris pada dataframe
# bilangan setelah koma untuk memilih kolom pada dataframe
```

```python
y = dataset[:,0]
y
```

```python
# Normalization
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# pisahkan data training dan testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
import numpy as np

Y_train = Y_train.astype(np.float32)
Y_test = Y_test.astype(np.float32)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
```

```python
model = Sequential([
                    Dense(32, activation='relu', input_shape=(5,)),
                    Dense(32, activation='relu', kernel_regularizer=l2(0.01)),
                    Dense(1, activation='sigmoid'),])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=150)
```

```python
model.evaluate(X_test, Y_test)
# elemen pertama adalah loss dan elemen kedua adalah akurasi
```
